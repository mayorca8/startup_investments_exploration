```python
# This program will analyze venture capiatl (VC) start-up investments
# To Begin, we will import the packages that will assist with the analysis 

import pandas as pd
from matplotlib import pyplot as plt
import math
import numpy as np
import seaborn as sb
```

```python
# Adding encoding to avoid utf-8 error

df_raw = pd.read_csv("investments_VC.csv", encoding = 'unicode_escape')

df_raw.head(10)
```

```python
df_raw.info()
# The data set has 54,294 rows and 39 columns, we see there are plenty null values
```

```python
# Removing rows where all values are null 

df = df_raw.drop('permalink', axis=1)
df = df.dropna(axis=0, how ='all')
df.head()
```

```python
# To reference columns easily, removing the white spaces to avoid issues down the road

df.columns = df.columns.str.replace(' ', '')
df.info()
```

```python
df['name'].is_unique
```

```python
# df['name'].duplicated(False)

df[df['name'].duplicated() == True]['name']
```

```python
# Cleaning the data and upating the data types 

df['name'] = df['name'].astype("string")
df['homepage_url'] = df['homepage_url'].astype("string")
df['market'] = df['market'].astype("category")
df['status'] = df['status'].astype("category")
df['country_code'] = df['country_code'].astype("string")
df['state_code'] = df['state_code'].astype("string")
df['region'] = df['region'].astype("string")
df['city'] = df['city'].astype("string")

# Below are all date related columns 
df['founded_at'] = df['founded_at'].astype("string")
df['founded_month'] = df['founded_month'].astype("string")
df['founded_quarter'] = df['founded_quarter'].astype("string")
df['first_funding_at'] = df['first_funding_at'].astype("string")
df['last_funding_at'] = df['last_funding_at'].astype("string")
```

```python
# Got an error when converting the funding_total_usd column to a float data type
# will clean up values in the columns to remove spaces

df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

df['funding_total_usd'] = df['funding_total_usd'].apply(lambda x: x.replace(',', ''))
df['funding_total_usd'] = df['funding_total_usd'].apply(lambda x: x.replace('-', ''))
# Will need to remove the commas and other special characters 
# from the funding total figures to avoid an error when attempting to convert to float 

df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'], errors='coerce')

df.info()
```

```python
# Statistical information across columns

print("These are the statistical values for the funding rounds per start up, not ignoring 0:")
print("---------------------\n")
print("The avg funding round count is ",df['funding_rounds'].mean())
print("The median funding round count is ",df['funding_rounds'].median())
print("The min funding round count is ",df['funding_rounds'].min())
print("The max funding round count is ",df['funding_rounds'].max())
print("The standard deviation for the seed investment is ",df['funding_rounds'].std())
print("---------------------\n")

```

```python
# To complete analysis by parts, we will create focused data frames 
# focusing on the top 5 markets

Software_df = df[df['market'] == ' Software ' ]
Biotechnology_df = df[df['market'] == ' Biotechnology ' ]
Mobile_df = df[df['market'] == ' Mobile ' ]
E_Commerce_df = df[df['market'] == ' E-Commerce ' ]
Curated_Web_df = df[df['market'] == ' Curated Web ' ]


# Will create a list to iterate through these categories

top_five_markets_by_count = ['Software', 'Biotechnology', 'Mobile', 'E-Commerce', 'Curated Web' ]
```

```python
# We will create a funciton that ouputs the visualizations
# We are interested in the avg financing by funding round to see if 
# there are obvious trends

def create_avg_vs_financing_round(market_df):
    """
    This function will complete work similar to that of the above for loop
    but this funciton should print graphs. This will allow for the creation
    of graphs by market. Assumes the data frames consist of the same shape and columns.
    """
    
    market_df_columns = list(market_df)
    
    market_averages = [] 
    
    for i in market_df_columns[30:]:
        
        # The avg will ignore rows with 0 since it implies no funding received
        # as opposed to funding received in the amount of 0
        i_mask = market_df[str(i)] != 0
        
        i_avg = market_df[i_mask][str(i)].mean()
            
        market_averages.append(i_avg)
        
        market_averages = [0 if math.isnan(x) else x for x in market_averages]
     
        
    print("The highest avg is: ", max(market_averages))

    print("-----------------\n")
    
    plt.bar(market_df_columns[30:], market_averages, color="#27db51", label="Avg Financing Amount by Round")
    plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
    plt.legend()
    plt.show()
    
    
    
```

```python
# Each market has its own story of innovation unfolding. Some markets are more ahead than others

print("The entire data set")
create_avg_vs_financing_round(df)

print("Market category: Software")
create_avg_vs_financing_round(Software_df)

print("Market category: Biotechnology")
create_avg_vs_financing_round(Biotechnology_df)

print("Market category: Mobile")
create_avg_vs_financing_round(Mobile_df)

print("Market category: E-Commerce")
create_avg_vs_financing_round(E_Commerce_df)

print("Market category: Curated Web")
create_avg_vs_financing_round(Curated_Web_df)
```

```python
# To create a comparative analysis across the top 5 markets, I will create a heatmap

funding_rounds = ['A', 'B', 'C', 'D', 'E', 'F','G', 'H']


def calc_avg(market_df):
    # This function takes a data frame 
    # returns an array of averages
    
    market_df_columns = list(market_df)
    
    market_averages = [] 
    
    for i in market_df_columns[30:]:
        
        i_mask = market_df[str(i)] != 0
        
        i_avg = market_df[i_mask][str(i)].mean()
        
        round(i_avg, 2)
            
        market_averages.append(i_avg)
        
        market_averages = [0 if math.isnan(x) else x for x in market_averages]
        
    return market_averages
    

# Creating an array as it will be needed for the heatmap
np.set_printoptions(suppress=True,
   formatter={'float_kind':'{:f}'.format})    
    
funding_rounds_avg = np.array([calc_avg(Software_df),
                              calc_avg(Biotechnology_df),
                              calc_avg(Mobile_df), 
                              calc_avg(E_Commerce_df),
                              calc_avg(Curated_Web_df)])


heat_map = sb.heatmap(funding_rounds_avg, xticklabels=funding_rounds, 
                      yticklabels=top_five_markets_by_count, cmap="YlGnBu")
plt.show()
```

```python

def print_status_pie(market_df):
    # This function will print out a pie chart broken down by status for
    # the provided data frame. Data frame must have a status column
    
    operating_mask = market_df['status'] == 'operating'
    acquired_mask = market_df['status'] == 'acquired'
    closed_mask = market_df['status'] == 'closed'
    
    operating_count = len(market_df[operating_mask])
    acquired_count = len(market_df[acquired_mask])
    closed_count = len(market_df[closed_mask])
    
    status_df = pd.DataFrame({'status':[operating_count, acquired_count, closed_count]},
                            index=['operating','acquired','closed'])
    
    plot = status_df.plot.pie(y='status', figsize=(5, 5), autopct='%1.0f%%')
    
    print("Total number of companies: ", len(market_df))
    print("--------\n")

```

```python
print("Market category: Software")
print_status_pie(Software_df)
```
